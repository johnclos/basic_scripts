```python
# https://www.geeksforgeeks.org/getting-all-csv-files-from-a-directory-using-python/
# importing the required modules
import glob
import pandas as pd
  
# specifying the path to csv files
path = "csvfoldergfg"
  
# csv files in the path
# files = glob.glob(path + "/*.csv")
files = glob.glob("*.csv")

# defining an empty list to store 
# content
data_frame = pd.DataFrame()
content = []
  
# checking all the csv files in the 
# specified path
for filename in files:
    
    # reading content of csv file
    # content.append(filename)
    df = pd.read_csv(filename, index_col=None)
    content.append(df)
  
# converting content to data frame
data_frame = pd.concat(content)
data_frame
```

```python
# write to csv
# https://sparkbyexamples.com/pandas/pandas-write-dataframe-to-csv-file/#write-pandas-dataframe-to-csv-file
data_frame.to_csv("weight.csv", header=True)
```
